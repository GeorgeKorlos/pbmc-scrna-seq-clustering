```python
import h5py
import numpy as np
import scipy.sparse

h5_path = "../data/raw/filtered_feature_bc_matrix.h5"
adata = sc.read_10x_h5(h5_path)

print("\n" + "="*60)
print("STATISTICS FROM filtered_feature_bc_matrix.h5")
print("="*60)

n_cells = adata.n_obs
print(f"\nExact number of barcodes (cells): {n_cells}")

n_genes = adata.n_vars
print(f"Exact number of features (genes): {n_genes}")

total_counts = np.sum(adata.X.data) if scipy.sparse.issparse(adata.X) else np.sum(adata.X)
print(f"Total non-zero counts: {int(total_counts)}")

umis_per_cell = np.asarray(adata.X.sum(axis=1)).flatten()
median_umis = np.median(umis_per_cell)
print(f"Median UMIs per cell: {median_umis:.0f}")

genes_per_cell = np.asarray((adata.X > 0).sum(axis=1)).flatten()
median_genes = np.median(genes_per_cell)
print(f"Median genes detected per cell: {median_genes:.0f}")

print(f"\nUMIs per cell distribution (percentiles):")
print(f"  10th percentile: {np.percentile(umis_per_cell, 10):.0f}")
print(f"  50th percentile: {np.percentile(umis_per_cell, 50):.0f}")
print(f"  90th percentile: {np.percentile(umis_per_cell, 90):.0f}")

print(f"\nGenes detected per cell distribution (percentiles):")
print(f"  10th percentile: {np.percentile(genes_per_cell, 10):.0f}")
print(f"  50th percentile: {np.percentile(genes_per_cell, 50):.0f}")
print(f"  90th percentile: {np.percentile(genes_per_cell, 90):.0f}")

print("\n" + "="*60)
```
